## 1. Proceso de carga y limpieza de datos.

**Cargamos librerias**

```python
import numpy as np
import pandas as pd
import os
```

```python
# Cargar archivo Excel

df = pd.read_excel("../data/raw/Adidas US Sales Datasets.xlsx",skiprows=4)

# Vista rápida de los primeros registros
df.head()
```

```python
# Borro la 1º column (Unnamed)
df = df.drop(df.columns[0], axis=1)
df.head()
```

**Exportamos la tabla tratada como un dataset**

```python
# Ruta destino con el nombre solicitado
output_path = "../data/processed/Adidas_US_Sales_procesed.csv"

# Exportar a CSV
df.to_csv(output_path, index=False)
```
